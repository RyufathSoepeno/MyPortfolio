# Problem 1: Plotting the Club Members

For this problem, you need to write the function **plotData1** to plot the data scientists in the data-science club so that we can conveniently visualize their tenured years, salaries, and paid accounts or not.  Your plot should be similar to the figure that can be displayed below by running the code cell.


```python
def plotData1(data):

    # Separate the data into paid and unpaid accounts
    
    # Paid accounts:
    # - We iterate over the 'data' list and check the third element (index 2) of each tuple.
    # - If the third element is 1, it indicates a paid account.
    # - We use list comprehension to extract the first element (index 0) as 'tenure' for paid members.
    # - We also extract the second element (index 1) as 'salary' for paid members.
    tenures_paid = [d[0] for d in data if d[2] == 1] 
    salaries_paid = [d[1] for d in data if d[2] == 1]  
    
    # Unpaid accounts:
    # - Similarly, we check the third element (index 2) of each tuple.
    # - If the third element is 0, it indicates an unpaid account.
    # - We again use list comprehension to extract the first element (index 0) as 'tenure' for unpaid members.
    # - We extract the second element (index 1) as 'salary' for unpaid members.
    tenures_unpaid = [d[0] for d in data if d[2] == 0] 
    salaries_unpaid = [d[1] for d in data if d[2] == 0]  


    plt.figure(figsize=(10.4, 7))

    plt.scatter(tenures_unpaid, salaries_unpaid,  marker = '+', color = 'blue', s = 50)
    plt.scatter(tenures_paid, salaries_paid, marker = 'o', color = 'mediumturquoise', edgecolor = 'black', s = 50)

    plt.legend(["unpaid", "paid"], loc="upper left")

    plt.xlim(0, 12)
    plt.xlabel("Years as Data Scientist", size=16)
    plt.ylabel("Salary", size=16)
    plt.title('Data-Science Club Members', size = 19)

    plt.show()
    pass

plotData1(data)
```

# Problem 3

For this problem, you need to write a function **plotData3** to make a figure with two subplots.  One is a bar plot, and the other is a histogram plot. The bar plot can conveniently support users to visualize how many accounts are paid and how many are unpaid. The specific numbers of accounts for each type should be presented. The histogram plot allows us discover the underlying frequency distribution of salary data associated with the two types of data scientists (paid and unpaid).  Your plot should be similar to the figure that can be displayed below by running the code cell.

```python
import matplotlib.pyplot as plt

def plotData3(data):
    # Separate data into paid and unpaid accounts
    paid_accounts = [d for d in data if d[2] == 1]
    unpaid_accounts = [d for d in data if d[2] == 0]
    
    # Use len to count the number of paid and unpaid accounts
    num_paid = len(paid_accounts)
    num_unpaid = len(unpaid_accounts)
    
    # Extract salaries for both paid and unpaid accounts
    paid_salaries = [d[1] for d in paid_accounts]
    unpaid_salaries = [d[1] for d in unpaid_accounts]

    #When you create more than one plot, you need to call out subplot

    # Create subplots with a vertical layout (2 rows, 1 column)
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9.5, 7))
    
    # Bar plot
    ax1.bar(['Unpaid', 'Paid'], [num_unpaid, num_paid], color=['blue', 'mediumturquoise'], edgecolor='black')
    ax1.set_title('Unpaid vs. Paid Accounts', size=18)
    ax1.set_ylabel('Number of Members', size=15)
    
    # Add legend to bar plot with total number
    paid_patch = mpatches.Patch(color = 'mediumturquoise', label=f'Paid ({num_paid}) accounts')
    unpaid_patch = mpatches.Patch(label=f'Unpaid ({num_unpaid}) accounts')
    
    # Add the legend to the bar plot, positioning it in the top-right corner
    ax1.legend(handles=[unpaid_patch, paid_patch])

    #Note: For some reason, I cannot have an edge color or a different color at the same time for "legend"


    
    # Histogram plot
    ax2.hist([unpaid_salaries, paid_salaries], bins=20, label=['Unpaid', 'Paid'], color=['blue', 'mediumturquoise'], alpha=0.7)
    ax2.set_title('Histogram of Salary', size=18)
    ax2.set_xlabel('Salary', size=15)
    ax2.set_ylabel('Number of Members', size=15)
    ax2.legend()

    # Display the plots
    plt.tight_layout()
    plt.show()

# Call the function to plot
plotData3(data)
```

# Problem 4 Community Data

For this problem, you need to write a function **barCommunityData** make a bar plot that shows numbers of counties in states.  Your plot should be similar to the figure that can be displayed below by running the code cell that has the Image object.  To approach this problem, you need to write the function definition that includes statements to parse the data from the provided file named "county.csv". For this assignment, you are {\bf not} allowed to use pandas library to process the data file.   You are required to read the file row by row and extract the data to address the question.  In your plot, the specific numbers of counties in the states should be presented. 

For the reference result plot, it shows the numbers of counties in Arizona, California, and Texas. However, your **barCommunityData** applications should allow any other states on the second argument.  

```python
import csv #important to process csv files

def barCommunityData(states):
    # Create a dictionary to store the count of counties per state
    state_counties = {state: 0 for state in states}
    
    # Open the CSV file--mode='r' means reading csv file
    with open('county.csv', mode='r') as file:
        reader = csv.reader(file)
        # Skip the header
        next(reader)
        
        # Iterate over each row in the file
        for row in reader:
            state = row[1]  
            if state in states:
                state_counties[state] += 1

    plt.figure(figsize=(10.8, 7))
    
    # Create the bar plot
    bars = plt.bar(state_counties.keys(), state_counties.values(), color=['blue', 'green', 'red', 'orange'])


    plt.title('Number of Counties in States', size=16)
    
    # Create a custom legend using patches
    handles = [plt.Rectangle((0, 0), 1, 1, color=bar.get_facecolor()) for bar in bars]
    plt.legend(handles, states, loc="upper left")
    plt.show()
    pass

barCommunityData(['Arizona', 'California', 'Texas'])

#Here below is the second argument attempt
barCommunityData(['Wisconsin', 'Arizona', 'Nebraska', 'Ohio'])
```
